```python
import json

with open('//a-filedata-p/Desktop/Neha.Sharma/Desktop/Final.json') as json_file:
    doc = json.load(json_file)
    
    #print (len(doc))

data = {'Viewport': [],'Page Event': [], 'Query String': [] ,'Title' :[],'PageName':[],'PostTime':[]}

for i in range (0,len(doc)):
    for transaction in [doc[i]]:
        data['Viewport'].append(transaction['data']['dom']['viewport_width'])
        #data['Login Status'].append(transaction['login_status'])
        if doc[i].get('data',{}).get('udo',{}).get('page_event',{}):
            data['Page Event'].append(transaction["data"]["udo"]["page_event"])
        else:
            data['Page Event'].append('')
        #(data['Page Event'].append(transaction["data"]["udo"]["page_event"]) if transaction["data"]["udo"]["page_event"] is not None else None)
        data['Query String'].append(transaction['data']['dom']['query_string'])
        data['Title'].append(transaction['data']['dom']['title'])
        
        if doc[i].get('data',{}).get('udo',{}).get('page_name',{}):
            data['PageName'].append(transaction['data']['udo']['page_name'])
        else:
            data['PageName'].append('')        
        
        data['PostTime'].append(transaction['post_time'])
    
print(data)
```

```python
import pandas
import pandas_profiling

out_df = pd.DataFrame.from_dict(data)
pandas_profiling.ProfileReport(out_df)
```
